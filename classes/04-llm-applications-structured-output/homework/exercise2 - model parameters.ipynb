# Exercise 2 - Model Parameters

Experiment with different model parameters to see how they affect the output.


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
```

```python
client = OpenAI(
    api_key=os.environ.get("GROQ_KEY"),
    base_url="https://api.groq.com/openai/v1",
)
```

```python
temperature = None
max_completion_tokens = None
system_prompt = (
    "You are a strict math tutor. Do not answer questions unrelated to Mathematics."
)
prompt = "What is the area of denmark"
top_p = None
```

```python
response = client.responses.create(
    input=prompt,
    model="openai/gpt-oss-20b",
    temperature=temperature,
    max_output_tokens=max_completion_tokens,
    instructions=system_prompt,
    top_p=top_p,
)
```
